```python
import csv
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection  import train_test_split
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn import svm
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix
```

```python
import pandas as pd
```

```python
df = pd.read_csv('features_data.csv')
```

```python
df
```

```python
features_data = df.sort('timestamp')
```

```python
features_data.head()
```

```python
features = features_data.drop(['Unnamed: 0', 'timestamp'], axis = 1)
```

```python
features.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 5)
pca.fit(features)
```

```python
print(pca.explained_variance_ratio_)
```

```python
features_new = pca.transform(features)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
plt.figure()
plt.scatter(features_new[:, 0], features_new[:,1], marker='x')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Cartesian coordinate PCA')
```

```python
features_df = pd.DataFrame(features_new)
```

```python
features_df.head()
```

```python
x_train, x_test = train_test_split(features_df, train_size = 0.2, random_state = 0)
```

```python
x_train.describe()
```

```python
clf = svm.OneClassSVM(kernel="rbf", gamma=0.1, nu = 0.15)
clf.fit(x_train)
```

```python
y_pred_train = clf.predict(x_train)
y_pred_test = clf.predict(x_test)
```
