```python
import pandas as pd
import numpy as np

data = pd.read_csv('character-predictions.csv')
data.head()
```

```python
data.shape
```

```python
#Dropping few features
new_data = data.drop(['dateOfBirth','DateoFdeath', 'mother', 'father', 'heir', 'spouse'], axis = 1)
```

```python
new_data.head()
```

```python
new_data.describe()
```

```python
new_data.drop(['isAliveMother', 'isAliveFather', 'isAliveHeir', 'isAliveSpouse', 'age'], axis = 1, inplace = True)
```

```python
new_data.head()
```

```python
char_name = new_data['name']
new_data.drop('name', axis = 1 , inplace = True)
```

```python
new_data.head()
```

```python
new_data.drop('isPopular', axis =1, inplace = True)
```

```python
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
new_data['numDeadRelations'] = scaler.fit_transform(new_data['numDeadRelations'])
```

```python
new_data.head()
```

```python
new_data.drop('boolDeadRelations', axis = 1, inplace= True)
```

```python
new_data.head()
```

```python
label_data = new_data['isAlive']
```

```python
features_data = new_data.drop('isAlive', axis = 1)
```

```python
features_data.head()
```

```python
features_data = pd.get_dummies(features_data)
```

```python
features_data.head()
```

```python
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel

lsvc = LinearSVC(C=0.1, penalty="l1", dual=False).fit(features_data, label_data)
model = SelectFromModel(lsvc, prefit=True)
X_new = model.transform(features_data)
```

```python
X_new.shape
```

```python
selected_features = model.get_support() 
```
